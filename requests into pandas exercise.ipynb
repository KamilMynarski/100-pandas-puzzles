```python
import pandas as pd
import numpy as np
import os
import requests
```

```python
def get_countries_by_language(language):
  url = 'https://restcountries.com/v3.1/lang/'+language
  try:
    response = requests.get(url)
    if response.status_code == 200:
      return pd.DataFrame(response.json())      
    elif response.status_code == 404:
      raise Exception(f"Language not found")
  except ZeroDivisionError:
    print("you can't divide by zero") 
  except Exception as e:
    print('error:', e)
```

```python
def transform_contries_table(df):
  df['official'] = df['name'].apply(lambda x: x.get('official'))
  df = df[['official'] + [c for c in df.columns if c != 'official']]
  df['capital'] = df['capital'].apply(lambda x: x[0])
  df['currency'] = df['currencies'].apply(lambda x: list(x.values())[0]['name'])
  df=df.drop(['cca2','status','name','idd','tld','ccn3','currencies','latlng','car','flags','coatOfArms','capitalInfo','postalCode','timezones'],axis=1)
  return df
```

```python
def save_into_file(df,language):
  #save file
  title = 'countries/' + language + '.csv'
  df.to_csv(title, index=False)
```

```python
def extract_countries_into_csv(language):
  df = get_countries_by_language(language)
  df = transform_contries_table(df)
  save_into_file(df,language)
```

```python
extract_countries_into_csv('Polish')
```
