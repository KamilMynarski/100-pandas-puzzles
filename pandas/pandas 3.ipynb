Filtering Data

```python
import pandas as pd

bios = pd.read_csv('./data/bios.csv')
coffee = pd.read_csv('./data/coffee.csv')
```

```python
bios.info()
```

```python
#to filter data by value of specific column and show all columns
bios.loc[bios['height_cm'] > 220]
#to show specific columns
bios.loc[bios['height_cm'] > 220,["name", 'height_cm']]
```

```python
#you dont have to use loc
bios[bios['height_cm'] > 220]
#to select columns in this case
bios[bios['height_cm'] > 220][['name','height_cm']]
```

```python
#to add another condition
bios[(bios['height_cm'] > 215) & (bios['born_country']=='USA')]
```

```python
#like condition
bios[bios['name'].str.contains('mateusz|kamil', case=False)]
#you can use regular expressions here: mateusz|kamil'
#to switch off regex use
bios[bios['name'].str.contains('mateusz', case=False, regex=False)]
```

```python
#there are other methods than 'contains'
bios[bios['born_country'].isin(['UKR','POL'])]
```

```python
#can combine methods
bios[bios['born_country'].isin(['GBR']) & (bios['name'].str.startswith('Keith'))]
```

```python
#other way to do this is to use 'query' function
bios.query('born_country == "USA" and born_city=="Seattle"')
```

```python
#adding/removing columns
coffee['price'] = 4.99
coffee.head()
```

```python
#conditions in new  column
import numpy as np

coffee['new_price'] = np.where(coffee['Coffee Type']=='Espresso',3.99,5.99)
coffee.head()
```

```python
#dropping column
coffee.drop(columns=['price'])
#it deesnt modify coffee df only returns new one
```

```python
#dropping column for called df
# coffee.drop(columns=['price'],inplace=True)
# its the same like this
# coffee = coffee.drop(columns=['price']
coffee.head()
```

```python
#pandas is storing one version of df in memory 
coffee_new = coffee
coffee_new['amount'] = 5
coffee.head()
```

```python
#to avoid it use copy function
coffee_new = coffee.copy()
coffee_new['storage'] = 1
coffee.head()
```

```python
#to select specific columns
coffee = coffee[['Day','Coffee Type','Units Sold','new_price']]
coffee.head()
```

```python
#you can have calculations in the column definition
coffee['revenue'] = coffee['Units Sold'] * coffee['new_price']
coffee.head()
```

```python
#to rename a column
coffee.rename(columns={'new_price':'price'},inplace=True)
coffee.head()
```
